## Question 12: Get team long name with total of goal, order the total number by descending

- Total of goal of a team is added up from both away and home games
- Use UNION statement 

```python
#write your query
query = """
SELECT team_long_name, 
       SUM(goal) AS total_goal FROM 
(
       SELECT team_long_name, SUM(M.home_team_goal) AS goal
  FROM Match M
  JOIN Team T
    On T.team_api_id = M.home_team_api_id
 GROUP BY T.team_long_name

 UNION ALL

       SELECT T.team_long_name, 
       SUM(M.away_team_goal) AS goal
  FROM Match M
  JOIN Team T
    On T.team_api_id = M.away_team_api_id
 GROUP BY T.team_long_name
)
 GROUP BY team_long_name
 ORDER BY total_goal DESC
"""

df = pd.read_sql(query, conn)
df
```

## Question 13: Get team long name with total of matches, order the total number by descending

```python
#write your query
query = """
SELECT team_long_name, 
       SUM(matches) AS matches FROM
(
       SELECT team_long_name, 
       COUNT(home_team_api_id) AS matches
  FROM Match M
  JOIN Team T
    On T.team_api_id = M.home_team_api_id
 GROUP BY T.team_long_name

 UNION ALL

       SELECT T.team_long_name, COUNT(away_team_api_id) AS matches
  FROM Match M
  JOIN Team T
    On T.team_api_id = M.away_team_api_id
 GROUP BY T.team_long_name
)
 GROUP BY team_long_name
 ORDER BY matches DESC
"""

df = pd.read_sql(query, conn)
df
```

## Question 14: Get numbers of win, lost and draw matches of each team

- If a team has home goal > away goal , team is â€œwinâ€ in this match
- If a team has home goal < away goal , team is â€œlostâ€ in this match
- If a team has home goal = away goal , team is â€œdrawâ€ in this match
- Use Case When statement


```python
#write your query
query = """ 
SELECT T.team_long_name,
       COUNT(CASE WHEN M.home_team_goal > M.away_team_goal THEN 'win' END) as win,
       COUNT(CASE WHEN M.home_team_goal < M.away_team_goal THEN 'lose' END) as lose,
       COUNT(CASE WHEN M.home_team_goal = M.away_team_goal THEN 'equa;' END) as equal
  FROM Match M
  JOIN Team T
    ON T.team_api_id = M.home_team_api_id
 GROUP BY T.team_long_name
 ORDER BY win DESC
"""
df = pd.read_sql(query, conn)
df
```

## Question 16: Get percentage of each league to join the matches

```python
#write your query
query = """ 
SELECT name as league, 
       ROUND(COUNT(*)*100.0 / sum(count(*)) over(),2) AS percentage_of_match
  FROM League, Match
 WHERE League.id = Match.league_id
 GROUP BY name
 ORDER BY percentage_of_match DESC
"""
df = pd.read_sql(query, conn)
df
```

## Question 17: Get percentage of score in each league

```python
#write your query
query = """ 
SELECT L.name as league, 
       ROUND(SUM(M.home_team_goal + M.away_team_goal)*100.0 / 
       SUM(SUM(M.home_team_goal + M.away_team_goal)) over(),2) AS percentage_of_goal
  FROM Match M
  JOIN League L
    ON L.id = M.league_id
 GROUP BY L.name
"""
df = pd.read_sql(query, conn)
df
```

## Question 18: Get total numbers of goals for each league in each season

## Question 19: Get player attributes

- Convert weight to kilogram
- Convert height to meter
- Calculuate bmi = ( (weight* 0.453592) / (height/100)^2)
- Get Age of player

```python
print(datetime.now())
#write your query
query = """ 
SELECT player_name, 
       weight*0.453592 as weight_kg, 
       height/100 as height_m, 
       (weight*0.453592)/((height/100)*(height/100)) as bmi, 
       (CURRENT_DATE - birthday) as years_old
  FROM Player
 ORDER BY years_old DESC
"""
df = pd.read_sql(query, conn)
df
```
